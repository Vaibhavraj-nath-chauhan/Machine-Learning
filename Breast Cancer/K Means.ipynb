```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt

bc = load_breast_cancer()

data = bc["data"]
target = bc["target"]
output = bc["target_names"]
```

#### K-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
sse = []
x = []
for i in range(1,10): 
    model = KMeans(n_clusters=i)
    model = model.fit(data)
    sse.append(model.inertia_)
    x.append(i)
    
plt.plot(x,sse)
plt.title("Eblow Method")
plt.show()
```

```python
model = KMeans(n_clusters=2)
model = model.fit(data)
labels = model.labels_
print("Accuracy Score: %.2f"%(accuracy_score(target,labels)*100))
```

```python
yes = []
no = []
for i in range(len(labels)):
    if labels[i]==target[i]:
        yes.append(1)
    else:
        no.append(0)
        
```

```python
plt.bar(yes,len(yes),color="g",width=.1)
plt.bar(no,len(no),color="r",width=.1)
plt.legend("YN")
plt.show()
```
